```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.info()
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Gender and Class")
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  
plt.figure(figsize=(10,5))
sns.barplot(x=df['FamilySize'], y=df['Survived'])
plt.title("Survival Rate by Family Size")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Embarked", y="Survived", data=df)
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df['Survived'] == 1]['Age'], bins=20, kde=True, label="Survived")
sns.histplot(df[df['Survived'] == 0]['Age'], bins=20, kde=True, label="Not Survived")
plt.legend()
plt.title("Survival Rate by Age")
plt.show()
```

```python
print("Key Patterns Found:")
print("- Women had higher survival rates, especially in higher classes.")
print("- Larger families had lower survival rates compared to individuals or small families.")
print("- Passengers from different embarkation ports had varying survival chances, possibly due to socio-economic status.")
print("- Younger passengers had higher survival rates compared to older ones.")
```
